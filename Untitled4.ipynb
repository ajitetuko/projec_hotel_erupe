```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Europe Hotel Booking Satisfaction Score.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
 #rata-rata usia pelanggan hotel
 average_age = df['Age'].mean()
 print(f'rata-rata usia pelanggan: {average_age:.2f} tahun')
```

```python
 #distribusi usia
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('distribusi usia')
plt.xlabel('jumlah')
plt.ylabel('usia')
plt.grid(True)
plt.show()
```

```python
#jenis kelamin
gender_distribution = df['Gender'].value_counts(normalize=True) * 100
print('jeis kelamin :')
print(gender_distribution)
```

```python
#visual data
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='purpose_of_travel', palette='viridis')
plt.title('tujuan perjalanan')
plt.xlabel('tujuan')
plt.ylabel('jumlah')
plt.grid(True)
plt.show()
```

```python
#jenis perjalanan 1
type_of_travel_distribution = df['Type of Travel'].value_counts(normalize=True) * 100
print('jenis Perjalanan:')
print(type_of_travel_distribution)
```

```python
#visual data 1
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Type of Travel', palette='viridis')
plt.title('distribusi jenis perjalan')
plt.xlabel('jenis perjalanan')
plt.ylabel('jumlah')
plt.grid(True)
plt.show()
```

```python
#nilai layanan hotel 1
service_columns = [
    'Hotel wifi service',
    'Hotel location',
    'Food and drink',
    'Stay comfort',
    'Common Room entertainment',
    'Checkin/Checkout service',
    'Other service',
    'Cleanliness'
    ]
average_scores = df[service_columns].mean()
print('skor layanan hotel:')
print(average_scores)
```

```python
#visual nilai rata2 layanan2
plt.figure(figsize=(15, 6))
average_scores.sort_values().plot(kind='barh', color='green')
plt.title('skor layanan hotel')
plt.xlabel('rata-rata')
plt.ylabel('layanan hotel')
plt.grid(True)
plt.show
```

```python
#nilai tiap layanan
plt.figure(figsize=(14, 10))
for i, column in enumerate(service_columns, 1):
  plt.subplot(4, 3, i)
  sns.histplot(df[column], bins=10, kde=True)
  plt.title(column)
  plt.xlabel('nilai')
  plt.ylabel('jumlah')
  plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#skor pemesanan online
average_online_booking_ease = df['Ease of Online booking'].mean()
print(f'nilai rata-rata kemudahan pemesanan hotel: {average_online_booking_ease:.2}')
```

```python
#rata-rata nilai kebersihan dan kenyamana
cleanliness_mean = df['Cleanliness'].mean()
stay_comfort_mean = df['Stay comfort'].mean()
print(f"rata-rata nilai kebersihan; {cleanliness_mean:.2f}")
print(f"rata-rata nilai kenyamanan; {stay_comfort_mean:.2f}")
```
